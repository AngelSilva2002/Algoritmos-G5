```python
import collections
```

```python

# Python program to demonstrate searching operation
# in binary search tree without recursion
class newNode:
# Constructor to create a new node
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

  # Function to check the given
  # key exist or not
def iterativeSearch(root, key):

    # Traverse until root reaches
    # to dead end
    string="La ruta del valor buscado desde el origen es:"
    while root != None:  
        # pass right subtree as new tree
        if key > root.data:
            root = root.right
            string+=" derecha"
        # pass left subtree as new tree
        elif key < root.data:
            root = root.left
            string+=" izquierda"
        else:
            return(string)
            # if the key is found return 1
            
    return "El número no se encuentra dentro del árbol"

# A utility function to insert a
# new Node with given key in BST
def insert(Node, data):

    # If the tree is empty, return
    # a new Node
    if Node == None:
      return newNode(data)

    # Otherwise, recur down the tree
    if data < Node.data:
      Node.left = insert(Node.left, data)
    elif data > Node.data:
      Node.right = insert(Node.right, data)

    # return the (unchanged) Node pointer
    return Node

#Función 1 para imprimir el árbol    
def printTree(root):
    levels=[]                                                     #O(1)
    cola = collections.deque()                                    #O(1)
    cola.append(root)                                             #O(1)
    while cola:                                                   #O(n)
      level = []                                                  #O(1)
      for i in range (len(cola)):                                 #O(n) 
        node = cola.popleft()                                     #O(1)
        if node:                                                  #O(1)
          level.append(node.data)                                 #O(1)
          cola.append(node.left)                                  #O(1)
          cola.append(node.right)                                 #O(1)
        else:                                                     #O(1)
          level.append(" ")                                       #O(1)
      if level: levels.append(level)                              #O(1)
    for level in levels:                                          #O(n)
      print(end=" "*int(2**(len(levels)-len(level))))             #O(1)
      for data in level:                                          #O(n)  
        print(data, end=" "*int(2**(len(levels)-len(level)))+" "*(len(str(data)))) #O(1) 
      print()                                                     #O(1)
       

  # Driver Code
if __name__ == '__main__':

  # Let us create following BST
  #    50
  # 30   70
  # / \ / \
  # 20 40 60 80
  root = None
  NodesNumber = int(input("Cantidad de nodos para la generacion del arbol binario: "))
  for i in range(NodesNumber):
    value = int(input("Ingrese el valor("+str(i+1)+"): "))
    root = insert(root, value)

  valueToSearch = int(input("Valor que desea buscar en el arbol: "))
  print(iterativeSearch(root, valueToSearch))
  print("El arbol ingresado es:")
  
  printTree(root)


# This code is contributed by PranchalK
```
